#     Machine Learning Code

```python
x=dd.iloc[:,0:7]
x
```

```python
y=dd.loc[:,['AQI_Bucket']]
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x.iloc[:,0]=le.fit_transform(x.iloc[:,0])
x.iloc[:,1]=le.fit_transform(x.iloc[:,1])
x.iloc[:,2]=le.fit_transform(x.iloc[:,2])
x.iloc[:,3]=le.fit_transform(x.iloc[:,3])
x.iloc[:,4]=le.fit_transform(x.iloc[:,4])
x.iloc[:,5]=le.fit_transform(x.iloc[:,5])
x.iloc[:,6]=le.fit_transform(x.iloc[:,6])

y=le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=0)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
pre=sv.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,pre))
print(confusion_matrix(y_test,pre))
```

```python
plt.plot(y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(n_estimators=50)
rc.fit(x_train,y_train)
prd=rc.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,prd))
print(confusion_matrix(y_test,prd))

```
